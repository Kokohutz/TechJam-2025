```python
import requests
import json

# The URL of the FastAPI application.
# If you are running it on your local machine, it will be http://127.0.0.1:8002
url = "http://127.0.0.1:8002/ocr/analyze"

# The headers to indicate that you are sending JSON data.
headers = {
    "Content-Type": "application/json"
}

# The data payload for the request.
# Replace the 'imageUrl' with the actual URL of the image you want to analyze.
data = {
    "imageUrl": "https://s3.ap-southeast-1.amazonaws.com/textract-public-assets-ap-southeast-1/DLRegular.png"
    # "imageUrl": "https://www.shutterstock.com/image-photo/man-holds-credit-card-he-600nw-2476576361.jpg"
}
# https://www.shutterstock.com/image-photo/man-holds-credit-card-he-600nw-2476576361.jpg

try:
    # Make the POST request
    response = requests.post(url, headers=headers, data=json.dumps(data))

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Print the JSON response from the server
        print("Analysis successful:")
        print(response.json())
    else:
        # Print the error status code and response text
        print(f"Error: {response.status_code}")
        print(response.text)

except requests.exceptions.RequestException as e:
    # Handle any exceptions that occur during the request
    print(f"An error occurred: {e}")
```

```python
def validate_text_with_service(text: str, validation_endpoint: str = "127.0.0.1:8003") -> Dict:
    """Validate text using the validation service"""
    try:
        headers = {"Content-Type": "application/json"}
        response = requests.get(f"http://{validation_endpoint}/validate_text_msg?text={text}", headers=headers)
        
        if response.status_code == 200:
            return response.json()
        else:
            print(f"Validation service failed with status code: {response.status_code}")
            return {
                'original_text': text,
                'encrypted_text': text,
                'entities': []
            }
    except requests.exceptions.RequestException as e:
        print(f"Error connecting to validation service: {e}")
        return {
            'original_text': text,
            'encrypted_text': text,
            'entities': []
        }
```
